```python
def mudar_loc(loc):
    new_loc = None
    if loc == 'America do Sul':
        new_loc = 'Brazil'
    else:
        new_loc = loc
    return new_loc

df['Location'] = df['Location'].apply(lambda e: mudar_loc(e))
```

```python
tag_cols = []

for col in df.columns.tolist():
    if 'TAGS.' in col:
        tag_cols.append(col)
        
tag_cols
```

```python
tag_cols = [col for col in df.columns.tolist() if 'TAGS.' in col]

tag_cols
```
